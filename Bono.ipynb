```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def r(A,B,omega,t):
    return A*np.exp(omega*t)-B*np.exp(-omega*t)
```

```python
def phi_f(omega,theta,t):
    return omega*t+theta
```

```python
def t(omega,theta,phi):
    return (phi-theta)/omega
```

```python
phi=np.linspace(-2*np.pi,2*np.pi,int(1e3))
t_prima=np.linspace(0,10,int(1e3))
omega=1e10#rad
theta=(np.pi)/2#iniciamos desde el 0 del sistema angular
A=1
B=1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(phi,r(A,B,omega,t(omega,theta,phi)))

plt.show()
```
